### Basic Graph

Marker reference: https://matplotlib.org/stable/gallery/lines_bars_and_markers/marker_reference.html


```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(8,5), dpi=100) # dots per inch, 100 pixels per inch

#8 * 100pxï¼Œ 5 * 100px

# Line 1

# Keyword Argument Notation
#plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

# Shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b^--', label='2x') # blue,  upward-pointing triangles, dashed line #any labels you want

## Line 2

# select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')

# Plot remainder of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--')

# Add a title (specify font parameters with fontdict) #this here is default for research papers
plt.title('Our First Graph!', fontdict={'fontname': 'Times New Roman', 'fontsize': 20}, loc = 'left') # loc can be 'left', 'center', and 'right'

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X, Y axis Tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4,], rotation=-45)
#plt.yticks([0,2,4,6,8,10])

# Add a legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
# plt.savefig('mygraph.png', dpi=300)

# Show plot
plt.show()
```

### Enhancement
fig, ax (Object-Oriented Interface)

```python
# enhance the quality of plots
"""
The â€˜retinaâ€™ setting only makes the plots look sharper 
in the notebook by increasing the display resolution (2*). 
It does not change the quality of images saved with plt.savefig() ->>>> When you present your results in a jupiter notebook
"""
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('retina')

labels = ['A', 'B', 'C']
values = [1, 4, 2]
colors = ['red', 'green', 'blue'] 

# Create figure and axis
# fig -> whole Figure
# ax -> Axes
fig, ax = plt.subplots(figsize=(5, 3), dpi=150)

# Create bar chart
bars = ax.bar(x=labels, height=values, color=colors)
# ax.bar(x=labels, height=values)
# ax.set_title("Bar Chart",fontdict={'fontname': 'Times New Roman', 'fontsize': 16})

# Border setting
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color("#1FC31D")

# Remove the ticks as well.
ax.tick_params(bottom=False, left=False)

# Add a horizontal grid (but keep the vertical grid hidden).
# Color the lines a light gray as well.
# ax.set_axisbelow(True)
# ax.yaxis.grid(True, color='#EEEEEE')
# ax.xaxis.grid(False)

# Grab the color of the bars so we can make the
# text the same color.
bar_color = bars[0].get_facecolor()

# Add text annotations to the top of the bars. -> Summme oben drauf
# Note, you'll have to adjust this slightly (the 0.3)
# with different data.
for bar in bars:
  ax.text(
      bar.get_x() + bar.get_width() / 2,
      bar.get_height() + 0.1,
      round(bar.get_height(), 1),
      horizontalalignment='center',
      color=bar_color,
      weight='bold'
  )


# Adjust layout to fit elements
fig.tight_layout()

# Save high-quality outputs
# fig.savefig('barchart.png', dpi=300)
# fig.savefig('barchart.jpg', quality=95)

plt.show()
```

__set_xticks()__ defines the positions of the ticks on the x-axis.
__set_xticklabels()__ assigns labels to those tick positions.

```python
# Sample Data
labels = ['A', 'B', 'C']
values = [1, 4, 2]

fig ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2,  figsize=(5, 3), dpi=150) #2 

fig, axs = plt.subplots(2, 2, figsize=(5, 3), dpi=150) #2 rows, 2 columns of subplots

axs[0, 0].plot([1, 2, 3], [4, 5, 6]) #oben links
axs[0, 0].set_title("Top-Left")

axs[0, 1].bar(labels, values) #oben rechts
axs[0, 1].set_title("Top-Right")

axs[1, 0].scatter([1, 2, 3], [4, 6, 5]) #unten links
axs[1, 0].set_title("Bottom-Left")
axs[1, 0].set_xlabel('Index')

# Rotate x-axis labels
axs[1, 0].set_xticks(range(len(labels)))  # Set tick positions
axs[1, 0].set_xticklabels(labels, rotation=-45)  # Set tick labels with rotation

axs[1, 1].hist([1, 2, 3, 4, 5])     #unten rechts
axs[1, 1].set_title("Bottom-Right")

plt.tight_layout()
plt.show()

```

__Seaborn:__ https://seaborn.pydata.org/tutorial/function_overview.html

```python
# real world example
#pip install seaborn
import seaborn as sns

def visualize_results(model, y_train, y_pred_train, y_test, y_pred_test, metrics, column, rs, color_train, color_test):
    # Create a DataFrame for seaborn
    train_data = pd.DataFrame({'Actual': y_train, 'Predicted': y_pred_train, 'Set': f"{column}_training"})
    test_data = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred_test, 'Set': f"{column}_testing"})
    plot_data = pd.concat([train_data, test_data])

    # Start plotting
    plt.figure(figsize=(10, 6))
    ax = sns.scatterplot(x='Actual', y='Predicted', hue='Set', data=plot_data, palette=[color_train, color_test], edgecolor='k')
    
    # Plot regression lines
    sns.regplot(x='Actual', y='Predicted', data=train_data, scatter=False, ax=ax, color=color_train, label=f'{column}_Training Line')
    sns.regplot(x='Actual', y='Predicted', data=test_data, scatter=False, ax=ax, color=color_test, label=f'{column}_Testing Line')

    # Set axes limits
    plt.xlim(0.0, 1.0)
    plt.ylim(0.0, 1.0)
    
    # Metrics Table
    table_data = [
        ["RÂ²", f"{metrics['r2_train']:.3f}", f"{metrics['r2_test']:.3f}"],
        ["RMSE", f"{metrics['rmse_train']:.3f}", f"{metrics['rmse_test']:.3f}"],
        ["RRMSE", f"{metrics['rrmse_train']:.3f}", f"{metrics['rrmse_test']:.3f}"]
    ]
    
    # Add table to the plot
    table = plt.table(cellText=table_data, colLabels=["Metric", "Training", "Testing"], loc='lower right', cellLoc='center', colWidths=[0.1, 0.1, 0.1])
    table.auto_set_font_size(False)
    table.set_fontsize(9)
    table.scale(1, 1.2)
    
    # Adjusting table properties and plot aesthetics
    for _, cell in table.get_celld().items():
        cell.set_edgecolor('lightgrey')
    
    plt.xticks(fontsize=12, fontname='Arial', fontweight='bold')
    plt.yticks(fontsize=12, fontname='Arial', fontweight='bold')
    plt.xlabel('Actual data', fontsize=14, fontname='Arial', fontweight='bold')
    plt.ylabel('Predicted data', fontsize=14, fontname='Arial', fontweight='bold')

    # Set legend position to upper left
    plt.legend(frameon=True, framealpha=1, edgecolor='lightgrey', loc='upper left')
    
    # Save the figure
    plt.savefig(f'results/{column}_{model.__class__.__name__.lower()}_{rs}.png', bbox_inches='tight')
```
