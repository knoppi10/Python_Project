```python
# change key and values
# values should be unique

# Original dictionary
original_dict = {
    'Canada': 'CAN',
    'United States of America': 'USA',
    'Mexico': 'MEX',
    'Brazil': 'BRA'
}

# Inverting the dictionary
inverted_dict = {value: key for key, value in original_dict.items()}

print(inverted_dict)


# Original dictionary with possible non-unique values
original_dict = {
    'Canada': 'CAN',
    'United States of America': 'USA',
    'Mexico': 'USA',  # Note the repeated value 'USA'
    'Brazil': 'BRA'
}

# Inverting the dictionary safely
inverted_dict = {}
for key, value in original_dict.items():
    if value not in inverted_dict:
        inverted_dict[value] = [key]
    else:
        inverted_dict[value].append(key)

print(inverted_dict)
```
